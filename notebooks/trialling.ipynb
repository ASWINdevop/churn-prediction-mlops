```python
import pandas as pd

# Load the split files
train = pd.read_csv("C:\\Users\\aswin\\Desktop\\Customer Churn Prediction\\data\\train.csv")
test = pd.read_csv("C:\\Users\\aswin\\Desktop\\Customer Churn Prediction\\data\\test.csv")

# Convert IDs to sets (mathematical collections of unique items)
train_ids = set(train['customerid'])
test_ids = set(test['customerid'])

# Find the intersection (items present in BOTH)
common_records = train_ids.intersection(test_ids)

print(f"Train Size: {len(train)}")
print(f"Test Size: {len(test)}")
print(f"Number of Common Records: {len(common_records)}")

if len(common_records) == 0:
    print("✅ INTEGRITY CHECK PASSED: No leakage detected.")
else:
    print(f"❌ CRITICAL FAILURE: Found {len(common_records)} duplicates!")
```
